```python
import numpy as np
import matplotlib.pyplot as plt
import os.path
from tabulate import tabulate #to export in table format
import sncosmo
```

```python
#filter lists
y_filter_i = []
z_filter_i = []

#reading the file
file = open("sdss_i.dat")
lines = file.readlines()
file.close()

for line in lines:
    if line.startswith('#'): continue
    co=line.rstrip().replace('INDEF','Nan').split()

    y_filter_i.append(co[0])
    z_filter_i.append(co[1])

#converting to float
y_filter_i = np.array(y_filter_i, dtype=float)
z_filter_i = np.array(z_filter_i, dtype=float)
```

```python
%matplotlib notebook
plt.plot(y_filter_i, z_filter_i, label="i")
plt.plot(y_filter_r, z_filter_r, label="r")
plt.plot(y_filter_g, z_filter_g, label="g")
plt.plot(y_filter_u, z_filter_u, label="u")
plt.legend(loc="upper right")
plt.show()
```
