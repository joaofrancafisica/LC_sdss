```python
import numpy as np
import matplotlib.pyplot as plt
import os.path
from tabulate import tabulate #to export in table format
```

```python
X_ = np.linspace(-10, 50, 60 + 1)
```

```python
#filter lists
M0_g = []
M1_g = []
M2_g = []

#reading the file
file = open("sdssg.dat")
lines = file.readlines()
file.close()

for line in lines:
    if line.startswith('#'): continue
    co=line.rstrip().replace('INDEF','Nan').split()

    M0_g.append(co[0])
    M1_g.append(co[1])
    M2_g.append(co[2])

#converting to float
M0_g = np.array(M0_g, dtype=float)
M1_g = np.array(M1_g, dtype=float)
M2_g = np.array(M2_g, dtype=float)
```

```python
#filter lists
M0_u = []
M1_u = []
M2_u = []

#reading the file
file = open("sdssu.dat")
lines = file.readlines()
file.close()

for line in lines:
    if line.startswith('#'): continue
    co=line.rstrip().replace('INDEF','Nan').split()

    M0_u.append(co[0])
    M1_u.append(co[1])
    M2_u.append(co[2])

#converting to float
M0_u = np.array(M0_u, dtype=float)
M1_u = np.array(M1_u, dtype=float)
M2_u = np.array(M2_u, dtype=float)
```

```python
#filter lists
M0_r = []
M1_r = []
M2_r = []

#reading the file
file = open("sdssr.dat")
lines = file.readlines()
file.close()

for line in lines:
    if line.startswith('#'): continue
    co=line.rstrip().replace('INDEF','Nan').split()

    M0_r.append(co[0])
    M1_r.append(co[1])
    M2_r.append(co[2])

#converting to float
M0_r = np.array(M0_r, dtype=float)
M1_r = np.array(M1_r, dtype=float)
M2_r = np.array(M2_r, dtype=float)
```

```python
#filter lists
M0_i = []
M1_i = []
M2_i = []

#reading the file
file = open("sdssi.dat")
lines = file.readlines()
file.close()

for line in lines:
    if line.startswith('#'): continue
    co=line.rstrip().replace('INDEF','Nan').split()

    M0_i.append(co[0])
    M1_i.append(co[1])
    M2_i.append(co[2])

#converting to float
M0_i = np.array(M0_i, dtype=float)
M1_i = np.array(M1_i, dtype=float)
M2_i = np.array(M2_i, dtype=float)
```

```python
%matplotlib notebook
plt.plot(X_, M0_r, label="r")
plt.plot(X_, M0_g, label="g")
plt.plot(X_, M0_u, label="u")
plt.plot(X_, M0_i, label="i")
plt.legend(loc="upper right")
plt.show()
```

```python
%matplotlib notebook
plt.plot(X_, M0_u, label="0")
plt.plot(X_, M1_u, label="1")
plt.plot(X_, M2_u, label="2")
plt.legend(loc="upper right")
plt.show()
```

```python
sumvec = []
for i in range(0,len(M0_i)):
    sumvec.append((1*(M0_i[i]+0*M1_i[i])))
```
